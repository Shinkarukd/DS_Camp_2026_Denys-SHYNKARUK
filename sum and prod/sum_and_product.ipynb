# Розв'язання загадки про Суму та Добуток

Існує два цілих числа $1 < a, b < 100$.
Вченому **S** повідомили суму цих чисел, а вченому **P** — їхній добуток.

**Діалог:**
1. **P**: "Я не знаю цих чисел."
2. **S**: "Я знав, що ти їх не знатимеш."
3. **P**: "Тоді я знаю ці числа."
4. **S**: "Тоді і я їх знаю."

### Крок 1: Підготовка простору варіантів
Спочатку згенеруємо всі можливі пари $(a, b)$ та згрупуємо їх за добутком і сумою. Це дозволить нам миттєво отримувати список кандидатів для будь-якого результату додавання чи множення.

```python
from collections import defaultdict

LIMIT = 100

pairs = [(a, b) for a in range(2, LIMIT) for b in range(a, LIMIT)]

sum_to_pairs = defaultdict(list)
prod_to_pairs = defaultdict(list)

for a, b in pairs:
    sum_to_pairs[a + b].append((a, b))
    prod_to_pairs[a * b].append((a, b))
```

### Крок 2: Аналіз перших двох реплік
**P** не знає чисел $\implies$ його добуток має більше одного варіанту розкладу на множники.
**S** каже, що знав це заздалегідь $\implies$ яку б пару чисел ми не взяли для суми, що є у **S**, їхній добуток **завжди** буде мати кілька варіантів розкладу.

```python

possible_sums = []

for s, candidates in sum_to_pairs.items():
    if all(len(prod_to_pairs[a * b]) > 1 for a, b in candidates):
        possible_sums.append(s)

print(f"Кількість потенційних сум: {len(possible_sums)}")
```

### Крок 3: Аналіз третьої репліки
**P** тепер знає відповідь $\implies$ він переглянув усі пари, що дають його добуток, і виявив, що лише **одна** з них має суму, яка входить до списку `possible_sums`.

```python

valid_products = []

for p, candidates in prod_to_pairs.items():
    matches = [pair for pair in candidates if (pair[0] + pair[1]) in possible_sums]
    
    if len(matches) == 1:
        valid_products.append(p)

print(f"Кількість 'прояснених' добутків: {len(valid_products)}")
```

### Крок 4: Фінальна репліка
**S** теж дізнається числа $\implies$ для його суми серед усіх можливих пар залишилася лише **одна**, чий добуток потрапив до списку `valid_products`.

```python
final_pair = None

for s in possible_sums:
    solutions = [(a, b) for a, b in sum_to_pairs[s] if (a * b) in valid_products]
    
    if len(solutions) == 1:
        final_pair = solutions[0]
        print(f"Результат знайдено: {final_pair[0]} та {final_pair[1]}")
```
