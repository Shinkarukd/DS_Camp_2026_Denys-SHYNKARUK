Task condition
Implement a solution to glasses dropping problem.

Given:

number of floors
number of glasses
Determine the minimum number of attempts required to find the critical floor (the highest floor from which a glass does not break), considering both outcomes of each drop:

glass breaks
glass does not break
Compute results for:

10 floors, 2 glasses
100 floors, 2 glasses
100 floors, 3 glasses
1000 floors, 3 glasses
1000 floors, 5 glasses

```python
def solve_glasses_problem(floors, glasses):
    """
    Визначає мінімальну кількість спроб для пошуку критичного поверху.
    """
    max_attempts = floors # У найгіршому випадку спроб стільки ж, скільки поверхів
    dp = [[0] * (max_attempts + 1) for _ in range(glasses + 1)]
    
    for n in range(1, max_attempts + 1):
        for k in range(1, glasses + 1):
            dp[k][n] = dp[k-1][n-1] + dp[k][n-1] + 1
        
        if dp[glasses][n] >= floors:
            return n
            
    return max_attempts

test_cases = [
    (10, 2),
    (100, 2),
    (100, 3),
    (1000, 3),
    (1000, 5)
]

print(f"{'Поверхи':<10} | {'Склянки':<10} | {'Мін. кількість спроб'}")
print("-" * 45)

for f, g in test_cases:
    attempts = solve_glasses_problem(f, g)
    print(f"{f:<10} | {g:<10} | {attempts}")
```
