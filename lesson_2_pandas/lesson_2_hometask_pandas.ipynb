```python
import pandas as pd 
import numpy as np 
```

# Question 1

```python
def answer_one():
    energy = pd.read_excel('Energy Indicators.xls', skiprows=17, skipfooter=38)
    energy = energy.drop(energy.columns[[0, 1]], axis=1)
    energy.columns = ['Country', 'Energy Supply', 'Energy Supply per Capita', '% Renewable']
    energy.replace('...', np.nan, inplace=True)
    energy['Energy Supply'] *= 1000000
    
    energy['Country'] = energy['Country'].str.replace(r"\d+", "", regex=True).str.replace(r"\(.*\)", "", regex=True).str.strip()
    energy['Country'] = energy['Country'].replace({
        "Republic of Korea": "South Korea", "United States of America": "United States",
        "United Kingdom of Great Britain and Northern Ireland": "United Kingdom",
        "China, Hong Kong Special Administrative Region": "Hong Kong"
    })

    # ВВП 
    gdp = pd.read_excel('world_bank.xls', sheet_name='Data', skiprows=3)
    
    gdp.rename(columns={gdp.columns[0]: 'Country'}, inplace=True)
    
    relevant_cols = ['Country']
    years_mapping = {} 
    
    target_range = range(2006, 2016)
    
    for col in gdp.columns:
        try:
            val = int(float(col))
            if val in target_range:
                relevant_cols.append(col)
                years_mapping[col] = str(val)
        except:
            continue
            
    gdp = gdp[relevant_cols]
    gdp.rename(columns=years_mapping, inplace=True)
    
    gdp['Country'] = gdp['Country'].replace({
        "Korea, Rep.": "South Korea", "Iran, Islamic Rep.": "Iran", "Hong Kong SAR, China": "Hong Kong"
    })

    # ScimEn 
    scimen = pd.read_excel('scimagojr.xlsx')
    scimen_top15 = scimen[:15]

    # Об'єднання
    df = pd.merge(scimen_top15, energy, how='inner', on='Country')
    df = pd.merge(df, gdp, how='inner', on='Country')
    
    df.set_index('Country', inplace=True)
    
    final_years = [str(y) for y in target_range]
    df[final_years] = df[final_years].apply(pd.to_numeric, errors='coerce')
    
    return df
```

```python
Top15 = answer_one()
Top15
```

# Questoin 2

```python
def answer_two():
    years = [str(y) for y in range(2006, 2016)]
    avgGDP = Top15[years].mean(axis=1).sort_values(ascending=False)
    return avgGDP.rename('avgGDP')

answer_two()
```

# Question 3

```python
def answer_three():
    target_country = answer_two().index[5]
    data = Top15.loc[target_country]
    return data['2015'] - data['2006']

answer_three()
```

# Question 4

```python
def answer_four():
    Top15['Ratio'] = Top15['Self-citations'] / Top15['Citations']
    max_ratio = Top15['Ratio'].max()
    country = Top15['Ratio'].idxmax()
    return (country, max_ratio)

answer_four()
```

# Question 5

```python
def answer_five():
    df = Top15.copy()
    df['PopEst'] = df['Energy Supply'] / df['Energy Supply per Capita']
    top3_country = df['PopEst'].sort_values(ascending=False).index[2]
    return str(top3_country)

answer_five()
```

# Question 6

```python
def answer_six():
    df = Top15.copy()
    df['PopEst'] = df['Energy Supply'] / df['Energy Supply per Capita']
    
    df['Citable docs per Capita'] = df['Citable documents'] / df['PopEst']
    
    correlation = df['Citable docs per Capita'].astype(float).corr(df['Energy Supply per Capita'].astype(float))
    return correlation

answer_six()
```

# Questoin 7

```python
def answer_seven():
    ContinentDict  = {'China':'Asia', 'United States':'North America', 'Japan':'Asia', 
                      'United Kingdom':'Europe', 'Russian Federation':'Europe', 'Canada':'North America', 
                      'Germany':'Europe', 'India':'Asia', 'France':'Europe', 'South Korea':'Asia', 
                      'Italy':'Europe', 'Spain':'Europe', 'Iran':'Asia', 'Australia':'Australia', 
                      'Brazil':'South America'}
    
    df = Top15.copy()

    df['PopEst'] = (df['Energy Supply'] / df['Energy Supply per Capita']).astype(float)
    
    df['Continent'] = df.index.to_series().map(ContinentDict)
    
    ans = df.groupby('Continent')['PopEst'].agg(['size', 'sum', 'mean', 'std'])
    
    return ans

answer_seven()
```
